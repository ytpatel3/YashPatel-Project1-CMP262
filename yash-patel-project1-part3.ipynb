# Question 1
## What events are most effective at producing computing degree students at CCM? (What events should CCM prioritize?) 

### Quantify the number of people that participated at events at CCM prior to enrolling
### count the yes values and get a sumn for each prior_particiapation column
### Show a bar chart sorted from highest to lowest

```python
# import Matplotlib for bar graph visualization
import matplotlib.pyplot as plt 

# variables that will store the number of people who participated in events prior to enrolling
open_house_participants = 0
instant_decision_day_participants = 0
on_campus_info_session_participants = 0
virtual_info_session_participants = 0
women_who_dare_participants = 0
regional_college_fair_participants = 0
hs_sharetime_info_session_participants = 0
challenger_program_participants = 0 
cybersec_program_participants = 0
hs_info_session_participants = 0
hs_campus_visit_participants = 0
individual_visit_participants = 0
workforce_dev_participants = 0
corp_training_participants = 0
teen_arts_fests_participants = 0
summer_camp_participants = 0

# use for loops to count number of "Yes"'s for each column; these counter variables will be displayed in the bar graph
for i in df['prior_participation_open_house']:
    if i == "Yes":
        open_house_participants += 1

for i in df['prior_participation_instant_decision_day']:
    if i == "Yes":
        instant_decision_day_participants += 1

for i in df['prior_participation_on_campus_info_session']:
    if i == "Yes":
        on_campus_info_session_participants += 1

for i in df['prior_participation_virtual_info_session']:
    if i == "Yes":
        virtual_info_session_participants += 1

for i in df['prior_participation_women_who_dare']:
    if i == "Yes":
        women_who_dare_participants += 1

for i in df['prior_participation_regional_college_fair']:
    if i == "Yes":
        regional_college_fair_participants += 1

for i in df['prior_participation_hs_sharetime_info_session']:
    if i == "Yes":
        hs_sharetime_info_session_participants += 1

for i in df['prior_participation_challenger_program']:
    if i == "Yes":
        challenger_program_participants += 1

for i in df['prior_participation_cybersec_program']:
    if i == "Yes":
        cybersec_program_participants += 1

for i in df['prior_participation_hs_info_session']:
    if i == "Yes":
        hs_info_session_participants += 1

for i in df['prior_participation_hs_campus_visit']:
    if i == "Yes":
        hs_campus_visit_participants += 1

for i in df['prior_participation_individual_campus_visit']:
    if i == "Yes":
        individual_visit_participants += 1

for i in df['prior_participation_workforce_development_class']:
    if i == "Yes":
        workforce_dev_participants += 1

for i in df['prior_participation_corporate_training']:
    if i == "Yes":
        corp_training_participants += 1

for i in df['prior_participation_teen_arts_fest']:
    if i == "Yes":
        teen_arts_fests_participants += 1

for i in df['prior_participation_CCM_summer_camp']:
    if i == "Yes":
        summer_camp_participants += 1

# dictionary with the data extracted above
data = {'Open House': open_house_participants , 'Instant Decision Day': instant_decision_day_participants, 'On Campus Info Session': on_campus_info_session_participants, 'Virtual Info Session': virtual_info_session_participants, 'Women Who Dare': women_who_dare_participants, 'Regional College Fair': regional_college_fair_participants, 'HS Sharetime Info Session': hs_sharetime_info_session_participants, 'Challenger Program':challenger_program_participants, 'Cybersec Program': cybersec_program_participants, 'HS Info Session': hs_info_session_participants, 'HS Campus Visit': hs_campus_visit_participants, 'Individual Campus Visit': individual_visit_participants, 'Workforce Development': workforce_dev_participants, 'Corporate Training': corp_training_participants, 'Teen Arts Fest': teen_arts_fests_participants, 'CCM Summer Camp': summer_camp_participants}
# data = {'1': open_house_participants , '2': instant_decision_day_participants, '3': on_campus_info_session_participants, '4': virtual_info_session_participants, '5': women_who_dare_participants, '6': regional_college_fair_participants, '7': hs_sharetime_info_session_participants, '8':challenger_program_participants, '9': cybersec_program_participants, '10': hs_info_session_participants, '11': hs_campus_visit_participants, '12': individual_visit_participants, '13': workforce_dev_participants, '14': corp_training_participants, '15': teen_arts_fests_participants, '16': summer_camp_participants}

# sort dictionary's values from highest to lowest
x = list(data.values())
x2 = sorted(x)

# x axis
events = list(data.keys())
# y axis
number_of = list(x2)

# size of bar graph
fig = plt.figure(figsize = (16, 9))
# create bar graph
plt.barh(events, number_of, color = 'seagreen')
plt.ylabel("Events", fontsize = 16)
plt.xlabel("Number of students", fontsize = 16)
plt.title("Events Students Have Attended Prior to Enrolling at CCM", fontsize = 24)
plt.show() 
```

## From the data above, we can answer question 1 by confidently saying that CCM summer camp is the most effective producer of computing degree students. The Teen Arts Festival is also an event that CCM can prioritize to improve computing degree enrollment. CCM can also advertise their rigourous courses so that employers can run corporate training/workforce development through CCM. 

# Question 2 
## What were the main sources from which the form's respondents received information about CCMâ€™s computing programs? (What is the most effective method to communicate with prospective students?)

```python
# variables that will store the number of people who got information from different sources
hs_guidance_counselor = 0
hs_teacher = 0
website = 0
admissions_office = 0
ccm_advisor = 0
employer = 0
workforce_dev = 0
NJ_workforce_dev = 0 
other = 0

# use for loops to count number of "Yes"'s for each column; these counter variables will be displayed in the bar graph
for i in df['info_source_hs_guidance_counselor']:
    if i == "Yes":
        hs_guidance_counselor += 1

for i in df['info_source_hs_teacher']:
    if i == "Yes":
        hs_teacher += 1

for i in df['info_source_ccm_website']:
    if i == "Yes":
        website += 1

for i in df['info_source_ccm_admissions']:
    if i == "Yes":
        admissions_office += 1

for i in df['info_source_ccm_advisor']:
    if i == "Yes":
        ccm_advisor += 1

for i in df['info_source_employer']:
    if i == "Yes":
        employer += 1

for i in df['info_source_workforce_development']:
    if i == "Yes":
        workforce_dev += 1

for i in df['info_source_NJ_workforce_development']:
    if i == "Yes":
        NJ_workforce_dev += 1

for i in df['info_source_other']:
    if i == "Yes":
        other += 1


# dictionary with the data extracted above
data2 = {'HS Guidance Counselor': hs_guidance_counselor , 'HS Teacher': hs_teacher, 'CCM Website': website, 'CCM Admissions Office': admissions_office, 'CCM Advisor': ccm_advisor, 'Employer': employer, 'Workforce Development': workforce_dev, 'NJ Workforce Development': NJ_workforce_dev, 'Other': other}

# sort dictionary's values from highest to lowest
vals = list(data2.values())
vals_sorted = sorted(vals)

# x axis
info_sources = list(data2.keys())
# y axis
num_of_info_sources = list(vals_sorted)

# size of bar graph
fig2 = plt.figure(figsize = (12, 6))
# create bar graph
plt.barh(info_sources, num_of_info_sources, color = 'pink')
plt.ylabel("Info Sources", fontsize = 16)
plt.xlabel("Number of Students Who Used Info Source", fontsize = 16)
plt.title("Students' Main Sources of Information of CCM's Computing Programs", fontsize = 24)
plt.show() 
```

## From the data above, we can answer question 2 by saying that CCM should revamp their advertisement strategies as CCM's advisors, admissions office, and website are significantly lower than other information sources.

## From the data above, we observe that students want to enroll in computing programs for their own personal enrichment. To motativate people to enroll in computing degrees, CCM can advertise their prestigious IT certification and their low tuition fees (second and third most popular motivations). Furthermore, they should emphasize other benefits like transferring to bachelor's degrees or possible job opportunities with an Associate's Degree (lowest rated motivations).

# Question 4
## Which computing degree program had the highest number of participants? (What computing degree program should CCM put more money into/advertise more of?)

```python
# variables that will store the number of people who got information from different sources
computer_science = 0
cis_gamedev = 0
it = 0
technical_studies = 0
data_certificate = 0
webdev_certificate = 0
digital_media_tech = 0
challenger_program = 0
sharetime_csip = 0

# use a for loop to count number of Degrees for each degree; these counter variables will be displayed in the pie chart
for i in df['current_degree']:
    if i == 'Computer Science':
        computer_science += 1
    if i == 'CIS Game Development Option':
        cis_gamedev += 1
    if i == 'Information Technology':
        it += 1
    if i == 'Technical Studies, Applied Associate of Science':
        technical_studies += 1
    if i == 'Data Analytics Certificate of Achievement':
        data_certificate += 1
    if i == 'Web Development Certificate of Achievement':
        webdev_certificate += 1
    if i == 'Digital Media Technology':
        digital_media_tech += 1
    if i == 'Challenger Program':
        challenger_program += 1
    if i == 'ShareTime CSIP Program':
        sharetime_csip += 1

# dictionary with the data extracted above
data4 = {'Computer Science': computer_science, 'CIS Game Development Option': cis_gamedev, 'Information Technology': it, 'Technical Studies, Applied Associate of Science': technical_studies, 'Data Analytics Certificate of Achievement': data_certificate, 'Web Development Certificate of Achievement': webdev_certificate, 'Digital Media Technology': digital_media_tech, 'Challenger Program': challenger_program, 'ShareTime CSIP Program': sharetime_csip}

# pie chart
plt.figure(figsize=(8, 8))
plt.pie(data4.values(), labels=data4.keys(), autopct='%1.1f%%')
plt.title('Distribution of CCM Computing Programs')
plt.show()
```

## From the data above, we observe that students enroll mostly in the Computer Science and Information Technology Majors. Since these majors are the most popular, CCM should prioritize funding and advertisting for these two majors.
